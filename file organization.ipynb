```python
from watchdog.observers import Observer
from watchdog.events import FileSystemEventHandler
from time import sleep
```

```python
folderpath = 'C:\\Users\\Essam\\Dropbox\\PC\\Downloads'
dir_music = os.path.join(folderpath, 'music')
dir_video = os.path.join(folderpath, 'video')
dir_image = os.path.join(folderpath, 'image')
dir_doc = os.path.join(folderpath, 'doc')
dir_exe = os.path.join(folderpath, 'exe')
dir_zip = os.path.join(folderpath, 'zip')
dir_other = os.path.join(folderpath, 'other')
dir_folder = os.path.join(folderpath, 'folder')

# Ensure directories exist
for directory in [dir_music, dir_video, dir_image, dir_doc, dir_exe, dir_zip, dir_other, dir_folder]:
    os.makedirs(directory, exist_ok=True)
```

```python

class FileMover(FileSystemEventHandler):
    def on_modified(self,event):
        for file in os.listdir(folderpath):
            if file.endswith('.mp3') or file.endswith('.wav'):
                move_file(file,dir_music)
            elif file.endswith('.mp4') or file.endswith('.avi'):
                move_file(file,dir_video)
            elif file.endswith('.jpg') or file.endswith('.png'):
                move_file(file,dir_image)   
            elif file.endswith('.doc') or file.endswith('.pdf'):
                move_file(file,dir_doc)
            elif file.endswith('.exe'):
                move_file(file,dir_exe)
            elif file.endswith('.zip'):
                move_file(file,dir_zip)
            elif file.endswith('.rar'):
                move_file(file,dir_zip)
            elif file.endswith('.pkt'):
                move_file(file,dir_exe)
            elif file.endswith('.txt'):
                move_file(file,dir_doc)
            elif file.endswith('.csv'):
                move_file(file,dir_doc)
            elif os.path.isdir(folderpath+'\\'+file):
                move_file(file,dir_folder)
            else:
                move_file(file,dir_other)
def move_file(file,dir):
    source = os.path.join(folderpath, file)
    destination = os.path.join(dir, file)
    if os.path.isdir(source) and dir == dir_folder:
        print(f"Skipping folder '{file}'")
        return 
    if os.path.isdir(source) and dir != dir_folder:
        print(f"Skipping folder '{file}'")
        return

    if os.path.exists(destination):
        print(f"Skipping file '{file}' as it already exists in '{dir}'")
        return    
    # these if condition was made to aviod moving the project file move than once
    os.rename(folderpath+'\\'+file,dir+'\\'+file)
    
for file in os.listdir(folderpath):
    if os.path.isdir(os.path.join(folderpath, file)) and file not in dir_folder:
        move_file(file, dir_folder)
    
    if file.endswith('.mp3') or file.endswith('.wav'):
        move_file(file,dir_music)
    elif file.endswith('.mp4') or file.endswith('.avi'):
        move_file(file,dir_video)
    elif file.endswith('.jpg') or file.endswith('.png'):
        move_file(file,dir_image)   
    elif file.endswith('.doc') or file.endswith('.pdf')or file.endswith('.docx')or file.endswith('.xlsx'):
        move_file(file,dir_doc)
    elif file.endswith('.exe'):
        move_file(file,dir_exe)
    elif file.endswith('.zip')or file.endswith('.rar')or file.endswith('.pkt'):
        move_file(file,dir_zip)
    elif file.endswith('.txt')or file.endswith('.csv'):
        move_file(file,dir_doc)
    elif file.endswith('.pka'):
        move_file(file,dir_exe)
    else:
        move_file(file,dir_other)

        



```

```python
if __name__ == "__main__":
    path = folderpath
    event_handler = FileMover()
    observer = Observer()
    observer.schedule(event_handler, path, recursive=True)
    observer.start()
    try:
        while True:
            sleep(10)
    except KeyboardInterrupt:
        observer.stop()
    observer.join()
```
