```python
import random

import string
import time as time
import os 

word_back = ["bicycle","lantern","whisper","mountain","jellyfish","puzzle","thunder","velvet","rocket","cactus","mirror","blanket","orchard"
             ,"spoon","pirate","tunnel","dragon","apricot","lantern"]

def get_word():
    word=random.choice(word_back)
    return word
def hide_word(word, revealed_letters=2):
    hidden = ['_' for _ in word]
    revealed_indices = random.sample(range(len(word)), revealed_letters)
    for idx in revealed_indices:
        hidden[idx] = word[idx]
    return hidden

def check_letter(letter, word, hidden_word):
    found = False
    if letter in word:
        for i in range(len(word)):
            if word[i] == letter and hidden_word[i] == '_':
               hidden_word[i] = letter
               found = True
    else:
        print("Sorry, that letter is not in the word.")
        return False
    return True
def check_win(hidden_word):
    if '_' not in hidden_word:
        print("Congratulations! You've guessed the word!")
        return True
    return False
def check_lose(tries):
    if tries <= 0:
        print("Sorry, you've run out of tries.")
        return True
    return False
def play_game():
    global tries
    word = get_word()
    hidden_word = list(hide_word(word))
    tries = 10
    print("Welcome to Hangman!")
    print("You have 10 tries to guess the word.")
    print("The word has", len(word), "letters.")
    print(" ".join(hidden_word))

    while True:
        letter = input("Guess a letter: ").lower()
        if len(letter) != 1 or not letter.isalpha():
            print("Please enter a single letter.")
            continue
        if not check_letter(letter, word, hidden_word):
            tries -= 1
            print("Tries left:", tries)
        print(" ".join(hidden_word))
        if check_win(hidden_word) or check_lose(tries):
            break
def main():
    while True:
        play_game()
        play_again = input("Do you want to play again? (y/n) ").lower()
        if play_again != 'y':
            print("Thanks for playing!")
            break
        
if __name__ == '__main__':
    main()
```
